```python
cross_tab = pd.crosstab(df['SeniorCitizen'], df['Churn'], normalize='index') * 100

# Plot the stacked bar chart
ax = cross_tab.plot(kind='bar', stacked=True, figsize=(4, 5), colormap='Set2')
plt.title("Percentage Distribution of Senior Citizens by Churn Status")
plt.xlabel("Senior Citizen")
plt.ylabel("Percentage")

# Add percentage labels on the bars
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    percentage = f'{height:.2f}%'
    
    if height > 0:  # Only add labels if the height is greater than 0
        ax.annotate(percentage, 
                    (x + width / 2, y + height / 2), 
                    ha='center', va='center',
                    fontsize=10, color='Black')

plt.legend(title="Churn", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# List of columns to plot
columns = ['PhoneService', 'MultipleLines', 'InternetService',
           'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
           'TechSupport', 'StreamingTV','StreamingMovies']

# Set the number of rows and columns for the subplot grid
n_cols = 3  # Number of columns per row
n_rows = -(-len(columns) // n_cols)  # Round up to get number of rows

# Set figure size dynamically based on grid
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 10), constrained_layout=True)

# Flatten axes array if needed
axes = axes.flatten()

# Plot countplot for each column
for i, col in enumerate(columns):
    sns.countplot(x=col, data=df, ax=axes[i], hue=df["Churn"], legend= True, palette='Set2')
    axes[i].set_title(f"Distribution of {col}")
    axes[i].set_xlabel(col)  
    axes[i].set_ylabel("Count")

# Hide any extra empty subplots
for i in range(len(columns), len(axes)):
    fig.delaxes(axes[i])

# Show plot
plt.show()
```

# The count plots indicate customer churn distribution across various telecom services. Customers without OnlineSecurity, TechSupport, and DeviceProtection show a higher churn rate. Fiber optic users tend to churn more than DSL users. Additionally, customers with "No internet service" generally have lower churn rates, suggesting that internet-based services influence retention.
